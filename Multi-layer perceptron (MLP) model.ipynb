#  Multi-layer perceptron (MLP)  Model

This dataset contains 500 rows of three values each. The values in the first and second column are explanatory variables/observations and the values in the third column are the corresponding response variable values.

The task of this notebook is to, 

Design a methodology to:

• choose an appropriate number of hidden nodes for an MLP applied to this regression problem.

• train an MLP with this number of hidden nodes and estimate its generalisation error.

And  produce graphs to help support my findings 

Importing libraries

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
```

Loading csv file into df varaible

```python
df=pd.read_csv('regression2.csv', sep=',',header=None) #No headers
df.head()
df
```

```python
X = df.iloc[:,0:2]
Y=df.iloc[:,2]
```

Spilting the data into training and testing data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.33,random_state=42)
reg_train = LinearRegression()
reg_train.fit(X_train, y_train)
```

Importing Axes3D from mpl_toolkits.mplot3d for 3-D plotting

```python
fig=plt.figure()
ax=Axes3D(fig)
ax.scatter(X.iloc[:,0],X.iloc[:,1:2],Y)

ax.set_xlabel(df.columns[0])
ax.set_ylabel(df.columns[1])
ax.set_zlabel(df.columns[2])
```

creating and using m1p1 varaible for MLPRegressor class 
Adding a single hidden layer to fit a model to the data
Importing mean_squared_error from from sklearn.metrics

```python
m1p1=MLPRegressor(hidden_layer_sizes=(1,))
m1p1.fit(X_train,y_train)
print('Model Loss:',m1p1.loss_)
pd.DataFrame(m1p1.loss_curve_).plot()
from sklearn.metrics import mean_squared_error as MS
pred=m1p1.predict(X_test)
print('MSE: {}'.format(MS(y_test,pred)))
#print('Intercept: {}'.format(reg.intercept_))
```

Adding Ten hidden layers and calculting loss

```python
m1p1=MLPRegressor(hidden_layer_sizes=(10,))
m1p1.fit(X_train,y_train)
print('Model Loss:',m1p1.loss_)
pd.DataFrame(m1p1.loss_curve_).plot()
```

one hundard hidden layers

```python
m1p1=MLPRegressor(hidden_layer_sizes=(100,))
m1p1.fit(X_train,y_train)
```

```python
print('Model Loss:',m1p1.loss_)
pd.DataFrame(m1p1.loss_curve_).plot()
```

Ninety Nine hidden layers

```python
m1p1=MLPRegressor(hidden_layer_sizes=(99,))
m1p1.fit(X_train,y_train)
print('Model Loss:',m1p1.loss_)
pd.DataFrame(m1p1.loss_curve_).plot()
```

Three Hundard Hidden layers

```python
m1p1=MLPRegressor(hidden_layer_sizes=(300,))
m1p1.fit(X_train,y_train)

print('Model Loss:',m1p1.loss_)
pd.DataFrame(m1p1.loss_curve_).plot()
```

Five hundred hidden layers

```python
m1p1=MLPRegressor(hidden_layer_sizes=(500,))
m1p1.fit(X_train,y_train)
print('Model Loss:',m1p1.loss_)
pd.DataFrame(m1p1.loss_curve_).plot()
```

```python
from sklearn.metrics import mean_squared_error as MS
pred=m1p1.predict(X_test)
print('MSE: {}'.format(MS(y_test,pred)))
```

```python
m1p1=MLPRegressor(hidden_layer_sizes=(100000,))
m1p1.fit(X_train,y_train)
```
